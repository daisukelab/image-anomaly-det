## Running manually

```python
# Train small epochs for x times
params.n_epochs = 1000 
N_TRAIN = 3

test_target='wood'

det = efficient_gan.EfficientGANAnoDet(params)

# Preprocess manually
dataroot = Path(dataroot)
det.reset_work(delete_all=True)

train_folder = det.work_folder/'train'
test_folder = det.work_folder/'test'

raw_train_files = sorted(dataroot.glob(f'{test_target}/train/good/*{params.suffix}'))
train_files = preprocess_images(raw_train_files, to_folder=train_folder, size=params.load_size,
                mode='RGB' if params.data.color else 'L', suffix=params.suffix, pre_crop_rect=params.data.pre_crop_rect,
                verbose = False)
preprocess_images
raw_test_files = sorted(dataroot.glob(f'{test_target}/test/*/*{params.suffix}'))
test_files = preprocess_images(raw_test_files, to_folder=test_folder, size=params.load_size,
                mode='RGB' if params.data.color else 'L', suffix=params.suffix, pre_crop_rect=params.data.pre_crop_rect,
                verbose = False)
test_labels = [f.name.split('/')[-1].split('_')[1] for f in test_files]
test_y_trues = [label != 'good' for label in test_labels]

# Train
det.prepare_experiment()
det.create_model()
det.setup_train(train_samples=train_files)
for i in range(N_TRAIN):
    det.train_model(train_samples=train_files)
    efficient_gan.visualize_train_and_generated(det.device, det.G, det.dl['val'], det.E)
    det.save_model(det.weights/f'{i}trained_weights')
    plt.show()

# Evaluate
for i in range(N_TRAIN):
    det.create_model(det.weights/f'{i}trained_weights')
    det.setup_runtime(ref_samples=test_files)
    values = det.evaluate_test(test_files, test_y_trues)
    print(values)
    efficient_gan.visualize_train_and_generated(det.device, det.G, det.dl['val'], det.E);
```
